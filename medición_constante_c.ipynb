Este experimento tuvo como objetivo medir la velocidad de la luz (o rapidez para los físicos estrictos) utilizando un método basado en la correlación de fotones entrelazados, comparando el resultado obtenido con el valor aceptado.

Se empleó un láser violeta de $405 nm$ y $120 mW$, incidiendo sobre un cristal de Beta-Bario-Borato ($\beta$BBO) tipo II para generar pares de fotones mediante el proceso Spontaneous Parametric Down-Conversion ($SPDC$). Se midió la diferencia de distancia entre los detectores y el cristal, así como la diferencia temporal entre las señales detectadas. Se realizaron 25 mediciones de distancia y 25 de tiempo, analizando únicamente los casos con coincidencias de fotones en ambos canales. 

# Fundamentos del experimento:

En un proceso de $SPDC$, un fotón de alta energía incidente en un cristal no lineal se divide espontáneamente en dos fotones de menor energía, conservando la energía y el momento total. 


Este proceso genera pares de fotones correlacionados, también conocidos como fotones entrelazados. El entrelazamiento cuántico es un fenómeno fundamental en la mecánica cuántica que describe una correlación fuerte entre partículas que persiste incluso cuando están separadas espacialmente.


La naturaleza correlacionada de los pares de fotones generados por $SPDC$ permite medir con gran precisión la diferencia en los tiempos de llegada a dos detectores separados espacialmente.

La velocidad de la luz ($c$) se puede calcular utilizando la relación:

$c = \frac{\Delta d}{\Delta t}$

Donde $\Delta d$ es la diferencia en las distancias recorridas por los fotones y $\Delta t$ es la diferencia en sus tiempos de llegada a los detectores.

Esta relación fundamental se deriva de la definición de velocidad en la mecánica clásica, pero se aplica igualmente a la propagación de la luz en el vacío.

Para detectar los fotones individuales, se utilizan detectores de fotones únicos, generalmente basados en fotodiodos de avalancha. Estos dispositivos son capaces de detectar niveles de luz extremadamente bajos, incluso fotones individuales, con una resolución temporal del orden de picosegundos.

El método utilizado en este experimento, aprovecha la precisión del SPDC y la alta resolución temporal de los osciloscopios al observar los pulsos emitidos por detectores de fotones únicos. Esta técnica permite realizar una medición precisa de la velocidad de la luz, con potenciales ventajas sobre métodos más tradicionales.

# Resultados:

Sobre $\Delta d$: Con respecto a este término en la definición de $v=\frac{\Delta d}{\Delta t}$, medimos veinticinco veces esta distancia asociada a los lentes con el flexómetro cuya mínima escala es de $1mm$. De aquí que cada medición tendrá una incertidumbre asociada de $\frac{1}{2}$(escala mínima), es decir, $0.5mm$.

Reportamos las siguientes mediciones en centímetros:

```python
import pandas as pd
# Creación de lista de valores medidos de delta d:
mediciones_delta_d = [108.9, 108.8, 108.9, 109.0, 108.9, 108.8, 108.7, 108.9, 
           108.8, 108.7, 108.9, 108.8, 108.9, 108.6, 108.9, 108.7, 
           108.9, 108.7, 108.8, 108.7, 108.8, 108.8, 109.0, 108.8, 109.0]

#Creación de un Data Frame para poder operar promedios:
delta_d = pd.DataFrame(mediciones_delta_d, columns=['Delta(d)'])
```

Con base en esas mediciones, calculamos el promedio de $\Delta d$ y su incertidumbre con respecto a las 25 mediciones:

```python
prom_delta_d = delta_d['Delta(d)'].mean()
print(prom_delta_d)
```

El promedio de $\Delta d$ es:

$\overline{\Delta d} =\frac{1}{25} \sum_{i=1}^{25} \Delta d _i 
    = 108.82800000000003 cm $
    

La incertidumbre de cada $\Delta d_i$ es $\sigma(\Delta d _i)=0.05cm$, por lo que tenemos que la propagación de incertidumbres está dada por:

$ \sigma (\overline{\Delta d})=\frac{1}{\sqrt{25}}\sqrt{\sum_{i=1}^{25} \sigma ^2 (\Delta d _i)}$

```python
#Hacemos el cálculo de la incertidumbre anterior:
import math
s_d = 0.05
n = 25
suma_s_d = n * s_d**2
incertidumbre_d = (1 / math.sqrt(n)) * math.sqrt(suma_s_d)
print(incertidumbre_d)
```

Por lo que tenemos que: $ \sigma (\overline{\Delta d})=0.05cm $

De lo anterior, se determina $\Delta d$ como:

$\Delta d = 1.08828 \pm 0.0005m$

Sobre $\Delta t$: Se presentan los resultados, análogos, pero con respecto al término temporal en nanosegundos:

```python
mediciones_delta_t = [4.53, 3.68, 4.68, 4.17, 4.28, 3.68, 4.31, 4.24,
                      4.46, 3.93, 4.41, 4.17, 4.50, 4.54, 4.55, 4.32,
                      4.34, 4.35, 4.24, 4.46, 4.46, 2.52, 4.69, 4.43, 4.53]

delta_t = pd.DataFrame(mediciones_delta_t, columns=['Delta(t)'])

prom_delta_t = delta_t['Delta(t)'].mean()
print(prom_delta_t)
```

De lo anterior, se determina $\Delta d$ como:

$ \overline{\Delta t}=\frac{1}{25} \sum_{i=1}^{25} \Delta t _i=4.2588ns $

En cuanto a la incertidumbre temporal, asociada a la diferencia de fase, se optó por partir de tomar la escala mínima del osciloscopio; esto debido a la dificultad de conocer todos los valores presentes en la fórmula asociada a la precisión de la medición de la delta de tiempo ($Delta$ $Time$ $Measurement$ $Accuracy$) que presenta el fabricante.

De acuerdo al manual registramos la escala mínima, $50ps$, por lo que tomamos la incertidumbre como $\frac{1}{2}$(escala mínima), es decir, $25ps$. 

$ \sigma (\overline{\Delta t})=\frac{1}{\sqrt{25}}\sqrt{\sum_{i=1}^{25} \sigma ^2 (\Delta t _i)}=\frac{1}{5} \sqrt{25(0.025)^2} =0.025ns $

Con ello se determinó $\Delta t$ como:

$\Delta t = 4.2588 \pm 0.025 ns$

Sobre $v$: De acuerdo a la ecuación $v=\frac{\Delta d}{\Delta t}$, usando los resultados reportados anteriormente tenemos que: 

$ v = \frac{\overline{\Delta d}}{\overline{\Delta t}}=\frac{108.82800000000003cm}{4.2588ns} $ 

```python
v=108.82800000000003/4.2588 #cm/ns
print(v)
```

Por lo que:

$ v=2.55536770921386x10^8 \frac{m}{s} $


Ahora, con respecto a la propagación de las incertidumbres asociadas a $v$, reportamos:

$\sigma (v)=v\cdot \sqrt{(\frac{ \sigma (\overline{\Delta d})}{\Delta d})^2+(\frac{ \sigma (\overline{\Delta t})}{\Delta t})^2}$

```python
#Retomamos y definimos las variables asociadas a la expresión anterior:
import math
D_d=108.828 #cm
D_t=4.2588 #ns
sig_D_d=0.05 #cm
sig_D_t=0.025 #ns
v=25.55367709213864 #cm/ns
```

```python
sigma_v=v*math.sqrt(((sig_D_d/D_d)**2)+(sig_D_t/D_t)**2)
print(sigma_v)
```

Obtenemos que $\sigma (v)=0.15046388784190784nm$, por lo que finalmente obtenemos $v$ aproximadamente como:

$v=2.555 \pm 0.015x10^8 \frac{m}{s}$

# Análisis de los errores:

De acuerdo a los resultados, con respecto a nuestra medición de $v$, reportamos el siguiente error relativo porcentual:

$ E_r ^v=\frac{\sigma (v)}{v} \cdot 100\% $

```python
E_rv=((sigma_v)/v)*100
print(E_rv)
```

$E_r ^v \approx 0.59 \%$

Ahora, en cuanto al error de nuestro cálculo con respecto a la definición de la velocidad real de la luz, $c\approx 29.979246\frac{cm}{ns} $, reportamos lo siguiente para el error absoluto:

$E_a =|valor_{real}-valor_{medido}|=|c-v|$

```python
c=29.979246
E_a=abs(c-v)
print(E_a)
```

$E_a \approx 4.425 \frac{cm}{ns}$

De lo anterior, para el error relativo porcentual de $v$ con respecto a $c$ tenemos:

$E_r =\frac{E_a}{valor_{real}}\cdot 100\%$

```python
E_r=(E_a / c)*100
print(E_r)
```

$E_r=14.76\%$

# Conclusión del análisis de errores:

Con una incertidumbre relativa del 0.59\% nuestra medida difiere en un $14.76\%$ del valor aceptado, lo cual se atribuye principalmente a errores sistemáticos en la medición de distancias y limitaciones en la resolución temporal del osciloscopio. Esto sugiere la necesidad de mejoras en el control de errores sistemáticos, y el uso de técnicas de medición de distancia más precisas y equipos con mayor resolución temporal.
